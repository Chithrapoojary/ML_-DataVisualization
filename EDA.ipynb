**Advanced visualization - Exploratory Data Analysis(EDA)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv("/content/tips.csv")
```

```python
data.head()
```

#Scatter Plot

```python
data.plot.scatter(x='total_bill',y='tip')
```

```python
data.plot.scatter(x='total_bill',y='tip',color='g')
```

```python
data.plot.scatter(y='sex',x='total_bill')
```

```python
data.plot.scatter(x='total_bill',y='day')
```

#Relational plot

```python
snr.relplot(x='total_bill',y='tip',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',style='smoker',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',style='smoker',col='day',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',style='day',col='smoker',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',style='smoker',col='time',data=data)
```

#Pair plot

```python
snr.pairplot(data)
```

#Cat plot

```python
snr.catplot(x='day',y='total_bill',hue='sex',data=data)
```

```python
snr.catplot(x='day',y='total_bill',hue='sex',kind='bar',data=data)
```

```python
snr.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=data)
```

**#Box plot**

```python
snr.catplot(x='day',y='total_bill',hue='sex',kind='box',data=data)
```
